```python
# Unique Values in Diabetes_binary columns
df['Diabetes_binary'].unique()
```

```python
# Assigning input features to 'X'
X = df.drop('Diabetes_binary', axis=1)
```

```python
# Assigning output to 'Y'
Y = df['Diabetes_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into training and test data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20)
```

```python
# training data shape
x_train.shape
```

```python
# testing data shape
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
feature_scale = StandardScaler()
feature_scale
```

```python
# Fitting and transforming the training data 
scaled_x_train = feature_scale.fit_transform(x_train)
```

```python
# Transforming the testing data using the parameters learnt from fitting the training data
scaled_x_test = feature_scale.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg_model = LogisticRegression()
log_reg_model
```

```python
# Training the Logistic Regression model
log_reg_model.fit(scaled_x_train, y_train)
```

```python
# Testing the model
y_pred_log_reg = log_reg_model.predict(scaled_x_test)
```

```python
# Outputs
y_pred_log_reg[:5]
```

```python
# Returns the probability of the output being in either of the class
y_pred_proba_log_reg = log_reg_model.predict_proba(scaled_x_test)
```

```python
# Probability that the output is '0' and '1'
y_pred_proba_log_reg[:5]
```

```python
from sklearn.metrics import log_loss, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
```

```python
log_loss(y_test, y_pred_proba_log_reg)
```

```python
confusion_matrix(y_test, y_pred_log_reg)
```

```python
accuracy_score(y_test, y_pred_log_reg)
```

```python
f1_score(y_test, y_pred_log_reg)
```

```python
precision_score(y_test, y_pred_log_reg)
```

```python
roc_auc_score(y_test, y_pred_log_reg)
```

```python
recall_score(y_test, y_pred_log_reg)
```

## Gradient Boosting Classifier

Now let's train a Gradient Boosting model and evaluate its performance.

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42, verbose=True)
gb_model
```

```python
# Training the Gradient Boosting model
gb_model.fit(scaled_x_train, y_train)
```

```python
# Making predictions with Gradient Boosting model
y_pred_gb = gb_model.predict(scaled_x_test)
y_pred_proba_gb = gb_model.predict_proba(scaled_x_test)
```

### Gradient Boosting Model Evaluation Metrics

```python
# Gradient Boosting Model Evaluation Metrics
print("=== GRADIENT BOOSTING CLASSIFIER EVALUATION METRICS ===")
print()

print(f"Accuracy Score: {accuracy_score(y_test, y_pred_gb):.4f}")
print(f"Precision Score: {precision_score(y_test, y_pred_gb):.4f}")
print(f"Recall Score: {recall_score(y_test, y_pred_gb):.4f}")
print(f"F1 Score: {f1_score(y_test, y_pred_gb):.4f}")
print(f"ROC AUC Score: {roc_auc_score(y_test, y_pred_gb):.4f}")
print(f"Log Loss: {log_loss(y_test, y_pred_proba_gb):.4f}")
print()

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_gb))
print()

print("Classification Report:")
print(classification_report(y_test, y_pred_gb))
```
